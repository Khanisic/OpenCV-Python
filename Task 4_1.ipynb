```python
import cv2 as cv
import numpy as np
```

```python
image = np.zeros((400, 400, 3))
```

```python
#creating a blank black colored image
blank=np.zeros((500,500,3),dtype='uint8') 

#creating a rectangle on left side vertically
blank=cv.rectangle(blank,(0,0),(250,500),(70,70,6),thickness=-1)

#creating a rectangle on right side vertically
blank=cv.rectangle(blank,(250,0),(500,500),(0,215,255),thickness=-1)

#drawing a big circle
blank=cv.circle(blank,(250,250),250,(255,255,255),thickness=-1)

#drawing a small circle
blank=cv.circle(blank,(250,250),125,(27,27,27),thickness=-1)

# Drawing an orange line top-left
cv.line(blank, (0,250), (250,0), (255,0,0), (5))

# Drawing an orange line bottom-right
cv.line(blank, (500,250), (250,500), (255,0,0), (5))

# Drawing an orange line top-right
cv.line(blank, (250,0), (500,250), (255,0,0), (5))

# Drawing an orange line bottom-right
cv.line (blank, (0,250), (250,500), (255,0,0), (5))

#drawing a smaller circle
cv.circle(blank,(250,250),50,(255,255,255),thickness=-1)
cv.imshow('Custom image', blank)
cv.waitKey(0) #to hold the photo
cv.destroyAllWindows() #to close the window
```

```python
img1 = cv.imread('ozil.jpg')
img2 = cv.imread('xavi.jpg')
print(img1.shape)
print(img2.shape)
```

```python
#extracting football image by cropping

football=img1[640:800,100:350]
cv.imshow('football cropped image', football)
cv.waitKey(0) #to hold the photo
cv.destroyAllWindows() #to close the window
```

```python
#extracting Barcelona Logo by cropping

logo=img2[1080:1450,1160:1600]
cv.imshow('cropped logo image', logo)
cv.waitKey(0) #to hold the photo
cv.destroyAllWindows() #to close the window
```

```python
#resizing football image
football= cv.resize(football, (50, 50))
```

```python
#swaping it with the football in the logo
logo[260:310,160:210]=football
cv.imshow('Swapped image', logo)
cv.waitKey(0) #to hold the photo
cv.destroyAllWindows() #to close the window
```

```python

'''                                                               
                             Merging two images using OpenCV with Python
'''
img1 = cv.imread('bottom.jpg')
img2 = cv.imread('top.jpg')
image = np.vstack((img2,img1))
image= cv.resize(image, (737, 491))
cv.imshow("Collage",image)
cv.waitKey(0)
cv.destroyAllWindows()
```
